**Tarea:**

1. Limpia el dato eliminando los espacios innecesarios en la variable `usuario_nombre`.

## Paso 9

Aplicar la limpieza a todos los registros garantiza consistencia antes de lanzar el programa de fidelización.

**Tarea:**

1. Elimina los espacios iniciales y finales de los nombres y reemplaza los guiones bajos por espacios.

````python
usuarios = [
    ['32415', ' mike_reed ', 32.0, ['ELECTRÓNICA', 'DEPORTE', 'LIBROS'], [894, 213, 173]],
    ['31980', 'kate morgan', 24.0, ['ROPA', 'LIBROS'], [439, 390]],
    ['32156', ' john doe ', 37.0, ['ELECTRÓNICA', 'HOGAR', 'COMIDA'], [459, 120, 99]],
    ]

usuarios_limpio = [] # se crea una lista nueva

usuario_nombre_1 = usuarios[0][1] # se accede al nombre del 1° usario: ' mike_reed '
usuario_nombre_1 = usuario_nombre_1.strip() # se eliminan los espacios, se devolverá: 'mike_reed'
usuario_nombre_1 = usuario_nombre_1.replace('_',' ') # se reemplaza el guíon bajo por un espacio:'mike reed'
usuario_nombre_1 = usuario_nombre_1.split() # se divide al 1° usuario en una lista
usuarios_limpio.append(usuario_nombre_1) # se agrega al 1° usuario en la lista usuarios_limpio
usuario_edad_1 = usuarios[0][2] # se accede a la edad del 1° usuario
usuario_edad_1 = int(usuario_edad_1) # se convierte la edad número entero
usuarios_limpio.append(usuario_edad_1) # se agrega la edad del 1° usuario en la lista usuarios_limpio

usuario_nombre_2 = usuarios[1][1] # se accede al nombre del 2° usario: 'kate morgan'
usuario_nombre_2 = usuario_nombre_2.split() # se divide al 2° usuario en una lista
usuarios_limpio.append(usuario_nombre_2) # se agrega al 2° usuario en la lista usuarios_limpio
usuario_edad_2 = usuarios[1][2] # se accede a la edad del 2° usuario
usuario_edad_2 = int(usuario_edad_2) # se convierte la edad número entero
usuarios_limpio.append(usuario_edad_2) # se agrega la edad del 2° usuario en la lista usuarios_limpio

usuario_nombre_3 = usuarios[2][1] # se accede al nombre del 3° usario: ' john doe '
usuario_nombre_3 = usuario_nombre_3.strip() # se eliminan los espacios, se devolverá: 'john doe'
usuario_nombre_3 = usuario_nombre_3.split() # se divide al 3° usuario en una lista
usuarios_limpio.append(usuario_nombre_3) # se agrega al 3° usuario en la lista usuarios_limpio
usuario_edad_3 = usuarios[2][2] # se accede a la edad del 3° usuario
usuario_edad_3 = int(usuario_edad_3) # se convierte la edad número entero
usuarios_limpio.append(usuario_edad_3) # se agrega la edad del 3° usuario en la lista usuarios_limpio

print(usuarios_limpio)
````
