### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

````python
total_pedidos_por_producto = order_products['product_id'].value_counts() # Total de veces que se pide cada producto
reordenes_por_producto = order_products[order_products['reordered'] == 1]['product_id'].value_counts() # Total de veces que se reordena cada producto (reordered = 1)
````

````python
# Crear DataFrame con ambos conteos
proporcion_reorden = pd.DataFrame({'total_pedidos': total_pedidos_por_producto, 'total_reordenes': reordenes_por_producto}).fillna(0)  # Llenar NaN con 0 para productos que nunca se reordenan
proporcion_reorden['tasa_reorden'] = proporcion_reorden['total_reordenes'] / proporcion_reorden['total_pedidos'] # Calcular la proporción
print("Primeras 10 tasas de reorden:")
proporcion_reorden.head(10)
````
